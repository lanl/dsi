```python
import sqlite3
```

```python
def execute_sql_file(sql_file: str, db_file: str):
    """Executes SQL commands from a file."""
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    with open(sql_file, "r") as f:
        sql_script = f.read()

    cursor.executescript(sql_script)
    conn.commit()
    conn.close()


def fetch_data(db_file):
    """Fetches and displays Customers and Orders tables."""
    conn = sqlite3.connect(db_file)
    cursor = conn.cursor()

    print("\nCustomers:")
    cursor.execute("SELECT * FROM Customers;")
    for row in cursor.fetchall():
        print(row)

    print("\nOrders:")
    cursor.execute("SELECT * FROM Orders;")
    for row in cursor.fetchall():
        print(row)

    conn.close()
```

```python
execute_sql_file("database/table_template.sql", "test.db")
fetch_data("test.db")
```
